```python
import json
from rapidfuzz import process, fuzz
```

```python
PRODUCT = {
    "barcode": None,
    "description": "CILINDRO MESTRE FREIO TRASEIRO TRILHA XRE300 2010 A 2018",
    "id": "3645",
    "ncm": "0092.38.47",
    "price": 10.5,
    "quantity": 10
}

with open('gweb_export_old.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

PRODUCTS_TO_MATCH = data['products']
```

```python
products_name_to_match = [p['name'] for p in PRODUCTS_TO_MATCH]
product_name_to_comparsion = PRODUCT['description']

match = process.extractOne(
    product_name_to_comparsion,
    products_name_to_match,
    scorer=fuzz.token_sort_ratio
)

print(match)
```
