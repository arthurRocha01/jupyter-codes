```python
import pandas as pd
import json
```

```python
df = pd.read_excel('clean_products_conversion.xlsx', dtype={
    'ID': int,
    'CÓD. BARRA': str,
    'DESCRIÇÃO': str,
    'QUANTIDADE': int,
    'PREÇO': float,
    'NCM': int
})
df.info()
```

```python
import pandas as pd

def transform_to_struct(row):
    old_id = int(row['ID'])
    name = str(row['DESCRIÇÃO'])
    quantity = int(row['QUANTIDADE'])
    price = float(row['PREÇO'])
    ncm = int(row['NCM'])
    barcode = str(row['CÓD. BARRA'])

    return {
        "old_id": old_id,
        "name": name,
        "product_type": "00",
        "measure_unit": "UNID",
        "conversion_factor": 1,
        "commission": 0,
        "ippt": "T",
        "obs": None,
        "has_serial_numbers": False,
        "active": True,
        "details": [
            {
                "old_id": old_id,
                "description": None,
                "current_quantity": quantity,
                "minimum_quantity": 0,
                "reserved_quantity": 0,
                "price": price,
                "wholesale_price": 0,
                "wholesale_quantity": 0,
                "cost_price": 0,
                "average_cost": 0,
                "profit_margin": 0,
                "net_weight": "0.0000",
                "gross_weight": "0.0000",
                "origin": 0,
                "ncm_code": ncm,
                "ex_ncm_code": None,
                "cest_code": None,
                "classification_code": None,
                "cest_ind_escala": None,
                "fci_code": None,
                "manufacturer_cnpj": None,
                "reference": None,
                "barcodes": [
                    {
                        "multiplier": 1,
                        "default": True,
                        "barcode": barcode
                    }
                ],
                "serial_numbers": []
            }
        ],
        "categories": []
    }
```

```python
json_list = df.apply(transform_to_struct, axis=1).tolist()
```

```python
output_data = {
    "owner": {
        "name": "ARTUR MOTO PECAS E ASSESSORIOS LTDA",
        "national_document": "49.488.758/0001-00"
    },
    "products": json_list
}

with open('gdoor_data_conversion.json', 'w', encoding='utf-8') as file:
    json.dump(output_data, file, ensure_ascii=False, indent=2, allow_nan=False)
```
