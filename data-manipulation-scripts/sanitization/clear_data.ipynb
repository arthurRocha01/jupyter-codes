```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('./products_database.xlsx', usecols=['CÓD. BARRA', 'DESCRIÇÃO', 'QUANTIDADE', 'PREÇO'] ,dtype=str)
df.info()
```

```python
df = df.dropna(how='all')
df.info()
```

```python
df = df.dropna(subset=['DESCRIÇÃO'])

df['DESCRIÇÃO'] = (
    df['DESCRIÇÃO']
    .str.strip()
    .str.upper()
)
df['CÓD. BARRA'] = df['CÓD. BARRA'].str.strip()

df['CÓD. BARRA'] = df['CÓD. BARRA'].fillna('')
df['QUANTIDADE'] = df['QUANTIDADE'].fillna('0')
df['PREÇO'] = df['PREÇO'].fillna('0')

df.info()
```

```python
df = df[df['DESCRIÇÃO'] != '']
df.info()
```

```python
df['QUANTIDADE'] = df['QUANTIDADE'].replace(r'^\s*$', np.nan, regex=True)
df['PREÇO'] = df['PREÇO'].replace(r'^\s*$', np.nan, regex=True)

df['QUANTIDADE'] = pd.to_numeric(df['QUANTIDADE'], errors='coerce').fillna(0).astype(int)
df['PREÇO'] = pd.to_numeric(df['PREÇO'], errors='coerce').fillna(0).astype(float)

df.info()
```

```python
df = df.reset_index(drop=True)
df.insert(0, 'ID', range(len(df)))

df.to_excel('../manipulation/clean_products_database.xlsx', index=False)
df.head()
```
