```python
import pandas as pd
from pathlib import Path
```

```python
folder = Path('./sheets')
files = list(folder.glob('*.xlsx'))
print(len(files))
```

```python
all_dfs = []
for file in files:
    df_temp = pd.read_excel(file, dtype=str)
    all_dfs.append(df_temp)

df = pd.concat(all_dfs, ignore_index=True)
df.info()
```

```python
df.to_excel('../sanitization/products_database.xlsx')
```

```python
for file in files:
    df = pd.read_excel(file, usecols=[0, 1, 2], dtype=str)
    df.to_excel(f'./processed_sheets/{file.name}')
```
