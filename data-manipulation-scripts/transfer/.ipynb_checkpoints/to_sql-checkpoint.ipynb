```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_excel('products_export.xlsx', dtype=str)
df.head()
```

```python
df = df.drop(columns=['current_quantity', 'wholesale_quantity', 'wholesale_price'], axis=1)
df['id_category'] = '1'
df.head()
```

```python
df.rename(columns={'barcode': 'sku'})
df.head()
```

```python
engine = create_engine('postgresql+psycopg2://userdev:admin@localhost:5432/api_java')

df = df.rename(columns={'name': 'description'})

df.to_sql(
    name='items',
    con=engine,
    schema='private',
    if_exists='append',
    index=False
)
```
