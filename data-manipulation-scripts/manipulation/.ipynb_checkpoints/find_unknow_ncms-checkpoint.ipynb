```python
import json
import pandas as pd
```

```python
with open('../gweb_export_old.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python
know_ncms = [
    '87141000', '40114000', '40139000', '84839000', '73151100',
    '85122000', '85113000', '85114000', '87149400', '94012000'
]
unknow_ncm = set()

for product in data['products']:
    for detail in product['details']:
        ncm = str(detail['ncm_code'])

        if ncm not in know_ncms:
            unknow_ncm.add((
                product['name'],
                ncm
            ))
```

```python
df = pd.DataFrame(list(unknow_ncm), columns=['Descrição', 'NCM'])
df.info()
```

```python
df.to_excel('unknow_ncms.xlsx', index=False)
```
