```python
import pandas as pd
```

```python
df = pd.read_excel('clean_products_export.xlsx', dtype=str)
df.head()
```

```python
df = df.drop(columns=['wholesale_price', 'wholesale_quantity'], axis=1)
df.head()
```

```python
import json
```

```python
def transform_to_struct(row):
    return {
        "old_id": 4593, # Dica: vocÃª pode usar row['id'] se tiver no CSV
        "name": row['name'], # Fechado o ]
        "product_type": "00",
        "measure_unit": "UNID",
        "conversion_factor": 1,
        "commission": 0,
        "ippt": "T",
        "obs": None, # Mudado de null para None
        "has_serial_numbers": False,
        "active": True,
        "details": [
            {
                "old_id": 4593,
                "description": None,
                "current_quantity": row['current_quantity'],
                "minimum_quantity": 0,
                "reserved_quantity": 0,
                "price": row['price'],
                "wholesale_price": 690,
                "wholesale_quantity": 2,
                "cost_price": 0,
                "average_cost": 0,
                "profit_margin": 0,
                "net_weight": "0.0000",
                "gross_weight": "0.0000",
                "origin": 0,
                "ncm_code": None,
                "ex_ncm_code": None,
                "cest_code": None,
                "classification_code": None,
                "cest_ind_escala": None,
                "fci_code": None,
                "manufacturer_cnpj": None,
                "reference": None,
                "barcodes": [
                    {
                        "multipler": 1,
                        "default": True,
                        "barcode": row['barcode']
                    }
                ],
                "serial_numbers": []
            }
        ],
        "categories": []
    }
```

```python
json_list = df.apply(transform_to_struct, axis=1).tolist()
```

```python
with open('gdoor_data_conversion_test.json', 'w', encoding='utf-8') as file:
    json.dump(json_list, file, ensure_ascii=False, indent=4)
```
