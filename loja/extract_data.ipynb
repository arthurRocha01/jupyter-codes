```python
import pandas as pd
import json
```

```python

with open('gweb_export.json', 'r') as file:
    data = json.load(file)

df = pd.json_normalize(
    data['products'],
    record_path=['details'],
    meta=['name'],
    errors='ignore'
    )

df.head()
# df.columns
```

```python
df['barcode'] = df['barcodes'].str[0].str.get('barcode')

df = df[[
    'name',
    'current_quantity',
    'price',
    'wholesale_price',
    'wholesale_quantity',
    'barcode'
]]


df.head()
```

```python
df.to_excel('products_export.xlsx', index=False)
```
