```python
import pandas as pd
from polyfuzz import PolyFuzz
```

```python
df = pd.read_excel('clean_products_export.xlsx', dtype=str)
df.head()
```

```python
df['name'] = df['name'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df.head()
```

```python
unique_names = df['name'].unique().tolist()

model = PolyFuzz('TF-IDF')
model.match(unique_names, unique_names)

df_matches = model.get_matches()

suggestions = df_matches[(df_matches['Similarity'] > 0.2) & (df_matches['Similarity'] < 1.00)]
suggestions = suggestions.sort_values(by='Similarity', ascending=False)
suggestions.head()
```
